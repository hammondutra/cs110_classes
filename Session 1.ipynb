```python
import matplotlib.pyplot as plt
import numpy as np

# Input sizes
n = np.linspace(1, 10, 100)

# Example functions for each notation
big_o = n**2  # O(n^2)
big_omega = n  # Ω(n)
big_theta = 5*n + 10  # Θ(n)

# Plot
plt.figure(figsize=(12, 8))

# Big-O
plt.subplot(3, 1, 1)
plt.plot(n, big_o, label="O(n²)", color="blue")
plt.title("Big-O Notation (Upper Bound)")
plt.xlabel("Input size (n)")
plt.ylabel("Steps / Time")
plt.legend()
plt.grid(True)

# Big-Omega
plt.subplot(3, 1, 2)
plt.plot(n, big_omega, label="Ω(n)", color="green")
plt.title("Big-Omega Notation (Lower Bound)")
plt.xlabel("Input size (n)")
plt.ylabel("Steps / Time")
plt.legend()
plt.grid(True)

# Big-Theta
plt.subplot(3, 1, 3)
plt.plot(n, big_theta, label="Θ(n)", color="red")
plt.title("Big-Theta Notation (Tight Bound)")
plt.xlabel("Input size (n)")
plt.ylabel("Steps / Time")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Input sizes
n = np.linspace(1, 25, 200)

# Actual function
f = 3*n**2 + 2*n + 7

# Bounds for comparison
big_o = 4*n**2          # O(n^2) (upper bound)
big_omega = 3*n**2      # Ω(n^2) (lower bound)
big_theta = 3.5*n**2    # Θ(n^2) (tight bound, example constant factor)

# Plot
plt.figure(figsize=(10, 6))

plt.plot(n, f, label="f(n) = 3n² + 2n + 7", color="black", linewidth=2)
plt.plot(n, big_o, label="O(n²) (Upper Bound)", linestyle="--", color="blue")
plt.plot(n, big_omega, label="Ω(n²) (Lower Bound)", linestyle="--", color="green")
plt.plot(n, big_theta, label="Θ(n²) (Tight Bound)", linestyle="--", color="red")

plt.title("Visualization of Big-O, Big-Omega, and Big-Theta for f(n) = 3n² + 2n + 7")
plt.xlabel("Input size (n)")
plt.ylabel("Steps / Time")
plt.legend()
plt.grid(True)
plt.show()
```
